Taylor Series Demonstration for the function e**x

```python
import matplotlib.pyplot as plt
import numpy as np
import math

# Config the matplotlib backend as plotting inline in IPython
%matplotlib inline

# Define three(3) term taylor series function for e**x for a X setpoint of zero (0)
def taylor_series_3(x):
    # Establish result as a floating point variable
    result = 0.0
    result = 1 + x + (x**2)/math.factorial(2)
    return result

# Define four(4) term taylor series function for e**x for a X setpoint of zero (0)
def taylor_series_4(x):
    # Establish result as a floating point variable
    result = 0.0
    result = 1 + x + (x**2)/math.factorial(2)+(x**3)/math.factorial(3)
    return result

#Establish np arra7 of x values from -5.0 to 5.0 with 0.25 interval step
x_values_np_array = np.arange(-5.0, 5.0, 0.25)
# Establish lists for e**x results
y_values_3_list = []
y_values_4_list = []
y_values_e_list = []

# Loop to compute e**x values
# Establish function input variable as floating point
function_input = 0.0
for i in range(0,len(x_values_np_array)):
    # Set function input value to x values in list
    function_input = x_values_np_array[i]
    # Compute three (3) function value and add to list
    y_values_3_list.append(taylor_series_3(function_input))
    # Compute four (4) function value and add to list
    y_values_4_list.append(taylor_series_4(function_input))
    # Compute e function value and add to list
    y_values_e_list.append(math.exp(function_input))

plt.plot(x_values_np_array, y_values_3_list, label='Three (3) Terms' )
plt.plot(x_values_np_array, y_values_4_list, label='Four (4) Terms')
plt.plot(x_values_np_array, y_values_e_list, label='e**x')
plt.title('e**x Taylor Series Demonstration in Python with X=0 Set Point')
plt.xlabel('X - Axis')
plt.ylabel('Y- Axis')
plt.legend()

```
